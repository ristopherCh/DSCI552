__1.(i) Fit a PCR model on the training set, with M (the number of principal components) chosen by cross-validation. Report the test error obtained.__  

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score

PCAx = range(1,122)
CrimePCACVScores = []
for i in PCAx:
    CrimePCA = PCA(n_components=i).fit_transform(CrimeTrainDFI) #or CrimeTrainDFIS for standardized
    CrimePCADF = pd.DataFrame(data = CrimePCA)
    CrimePCR = LinearRegression().fit(CrimePCADF, CrimeTrainTarget)
    CrimeCVScores = cross_val_score(CrimePCR, CrimePCADF, CrimeTrainTarget, cv=5, scoring = 'neg_mean_squared_error')
    CrimeCVScores = 0 - CrimeCVScores
    CrimePCACVScores.append(CrimeCVScores.mean())
CrimePCACVScores
plt.plot(PCAx, CrimePCACVScores)
```

```python
PCAmin = min(CrimePCACVScores)
PCAmin = round(PCAmin, 5)
PCAminindex = CrimePCACVScores.index(min(CrimePCACVScores))
print('From using cross-validation to find the M value that gives the lowest PCR score, the M value is {0} and the corresponding MSE is {1}.'.format(PCAminindex, PCAmin))
```

```python
CrimePCACV = PCA(n_components = PCAminindex)
CrimeTrainDFIPCACV = CrimePCACV.fit_transform(CrimeTrainDFI) #fit & transform the PCA to the training data, with CV chosen M
CrimeTestDFIPCACV = CrimePCACV.transform(CrimeTestDFI) #transform the test data with the PCA trained to the training data
CrimePCRTest = LinearRegression().fit(CrimeTrainDFIPCACV, CrimeTrainTarget) #Fit the linear regression to the training data
CrimeTestPredPCA = CrimePCRTest.predict(CrimeTestDFIPCACV)
CrimeTestPCAMSE = mean_squared_error(CrimeTestTarget, CrimeTestPredPCA)
CrimeTestPCAMSE = round(CrimeTestPCAMSE, 5)
print('The MSE on the testing data using CV determined M value for PCR is {0}.'.format(CrimeTestPCAMSE))
```
