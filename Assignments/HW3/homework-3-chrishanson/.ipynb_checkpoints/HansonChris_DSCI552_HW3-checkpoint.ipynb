Use python's bootstrapped or any other method to build a 90% bootstrap confidence interval for the standard deviation of each feature:

```python
#Here I "Bootstrap" my data - I sample it with replacement, and calculate stds of each sample, 1000 times, and store these stds

df_stds = pd.DataFrame()
for i in range(0,1000):
    asdf = df_allmetrics.sample(frac=1, replace=True)
    fdsa = asdf.std()
    df_stds = pd.concat([df_stds, fdsa], axis=1)
```

```python
#Here I calculate a 90% confidence interval on all the std values I Bootstrapped previously

means = df_stds.mean(axis=1)
stds = df_stds.std(axis=1)
sqsamp = math.sqrt(1000)
invsqsamp = 1/sqsamp
Z = 1.645 #a 90% confidence interval is associated with a Z value of 1.645
Zstds = stds.multiply(Z)
ZNstds = Zstds.multiply(invsqsamp)
low90 = means.sub(ZNstds)
high90 = means.add(ZNstds)
conf90 = pd.concat([low90, high90], axis=1)
conf90.columns = ['90% confidence lower', '90% confidence upper']
conf90.head()
```

Below I break up my training and set into (1,2,...,20) time series.

```python
#Here I split my training dataset in p cuts and put it side by side, in each DataFrame

length = len(dftraining_list)
df_pscan_list = []
df_pscanT = pd.DataFrame()
df_pscanTT = pd.DataFrame()
for i in range(0,length):
    df_x = dftraining_list[i]
    df_x = df_x.drop(columns=['# Columns: time'])
    for p in range (1,21):
        Q = round(480/p)
        for w in range(1,p+1):
            df_pscanT = df_x.iloc[Q*(p-(p-w+1)):Q*(p-(p-w))]
            df_pscanT = df_pscanT.reset_index(drop=True)
            df_pscanTT = pd.concat([df_pscanTT, df_pscanT], axis=1)
            df_pscanTT = df_pscanTT.dropna()
            
        df_pscan_list.append(df_pscanTT)
        df_pscanTT = pd.DataFrame()

#output: df_pscan_list is a list 1380 long. It is used to make 20 dataframes of 69 rows below.
```

```python
#Here I read out my split up training time series and convert it into dataframes of usable data

rfelist_total = []
df_pscanmetrics = pd.DataFrame()

for i in range(0,20): #should be (0,20)
    r = 6*i + 6
    allrows_pscan = np.zeros([69,3*r])

    for u in range(0,69):
        temprows = []
        w = i+20*u
        fds = df_pscan_list[w].describe()
        fds = fds.loc[['mean', '50%', 'std']]
        fds = fds.transpose()
        
        for y in range (0,r):
            for t in range(0,3):
                q = fds.iloc[y,t]
                temprows.append(q)
        
        allrows_pscan[u] = temprows
        temprows = []
        df_pscanmetrics = pd.DataFrame(data = allrows_pscan)
    
    rfelist_total.append(df_pscanmetrics)
    df_pscanmetrics = pd.DataFrame()

#output: rfelist_total, a list of my p = 20 training dataframes
```

__2. ISLR 3.7.4__  

(a) The training RSS for the linear regression would be small, but the training RSS for the cubic regression would be even smaller, as the extra terms would allow the cubic regression to fit itself closer to the inherent variability in the datapoints - the irreducible error.

(b) The test RSS for the linear regression would be smaller than the test RSS for the cubic regression. This is because the cubic regression would have likely "overfit" itself to the training data, following the trends of the epsilon values - which would have increased variance without decreasing bias. The linear regression would not have overfit.

(c) Again, when fitting to training data, the cubic regression will likely have the ability to fit itself closer to the datapoints than the linear model can, so it will have a lower training RSS. This is true whether the model is truly linear or moderately unlinear.

(d) In this rather ambiguously described case, it cannot be known whether the linear or the cubic regression would be more accurate, and result in a lower RSS. If the true relationship is nearly linear, the cubic regression could overfit and result in a higher test RSS, but if the true relationship is far from linear, the linear regression would introduce extra bias, resulting in higher test RSS.
