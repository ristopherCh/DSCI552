__(f)__ Nonlinear responses.

```python
#AT. There is evidence of a nonlinear response, according to the P values of the X^2 and X^3 values.
AT_df = pd.DataFrame(AT)
PolyReg = PolynomialFeatures(degree=3, include_bias=False)
PolyReg.fit_transform(AT_df)
#extract the X^2 and X^3 terms and add it to the edited df
AT_df['AT^2'] = PolyReg.fit_transform(AT_df[['AT']] )[:,1]
AT_df['AT^3'] = PolyReg.fit_transform(AT_df[['AT']] )[:,2]
#Polynomial Regression for AT
AT_df2 = sm.add_constant(AT_df)
Poly_AT = sm.OLS(y,AT_df2)
Poly_AT_results = Poly_AT.fit()
Poly_AT_results.summary()
```

```python
#V. There is evidence of a nonlinear response, according to the P value of the X^3 value.
V_df = pd.DataFrame(V)
PolyReg.fit_transform(V_df)
#extract the X^2 and X^3 terms and add it to the edited df
V_df['V^2'] = PolyReg.fit_transform(V_df[['V']] )[:,1]
V_df['V^3'] = PolyReg.fit_transform(V_df[['V']] )[:,2]
#Polynomial Regression for V
V_df2 = sm.add_constant(V_df)
Poly_V = sm.OLS(y,V_df2)
Poly_V_results = Poly_V.fit()
Poly_V_results.summary()
```

```python
#AP. There is evidence of a nonlinear response, according to the P values of the X^2 and X^3 values.
AP_df = pd.DataFrame(AP)
#PolyReg = PolynomialFeatures(degree=3, include_bias=False)
PolyReg.fit_transform(AP_df)
#extract the X^2 and X^3 terms and add it to the edited df
AP_df['AP^2'] = PolyReg.fit_transform(AP_df[['AP']] )[:,1]
AP_df['AP^3'] = PolyReg.fit_transform(AP_df[['AP']] )[:,2]
#Polynomial Regression for AP
AP_df2 = sm.add_constant(AP_df)
Poly_AP = sm.OLS(y,AP_df2)
Poly_AP_results = Poly_AP.fit()
Poly_AP_results.summary()
```

```python
#RH. There is evidence of a nonlinear response, according to the P values of the X^2 and X^3 values.
RH_df = pd.DataFrame(RH)
PolyReg.fit_transform(RH_df)
#extract the X^2 and X^3 terms and add it to the edited df
RH_df['RH^2'] = PolyReg.fit_transform(RH_df[['RH']] )[:,1]
RH_df['RH^3'] = PolyReg.fit_transform(RH_df[['RH']] )[:,2]
#Polynomial Regression for RH
RH_df2 = sm.add_constant(RH_df)
Poly_RH = sm.OLS(y,RH_df2)
Poly_RH_results = Poly_RH.fit()
Poly_RH_results.summary()
```

__(g)__ Are any interaction terms statistically significant? Yes.

```python
#Adding interaction terms to my X DataFrame (with constant)
All_X_int = All_X.copy(deep=True)
All_X_int["ATV"] = All_X_int['AT'] * All_X_int['V']
All_X_int["ATAP"] = All_X_int['AT'] * All_X_int['AP']
All_X_int["ATRH"] = All_X_int['AT'] * All_X_int['RH']
All_X_int["VAP"] = All_X_int['V'] * All_X_int['AP']
All_X_int["VRH"] = All_X_int['V'] * All_X_int['RH']
All_X_int["APRH"] = All_X_int['AP'] * All_X_int['RH']
All_X_int
```

```python
#Polynomial Regression for All X and interaction terms
All_X_int_reg = sm.OLS(y,All_X_int)
All_X_int_reg_results = All_X_int_reg.fit()
All_X_int_reg_results.summary()
```

The above shows that there is evidence of association of interactions of predictors with the response. The statistically significant terms are AT*V, AT*RH, V*AP, and AP*RH.

__(i)__ KNN Regression:
Perform k-nearest neighbor regression for this dataset using both normalized
and raw features. Find the value of k is a member of the set {1; 2; : : : ; 100} that gives you the
best fit. Plot the train and test errors in terms of 1/k.

```python
#Normalize All_X
All_X_2 = All_X.copy(deep=True)
All_X_2 = All_X_2.drop(columns=['const'])
All_X_norm = (All_X_2 - All_X_2.min())/(All_X_2.max()-All_X_2.min())
```

```python
#KNN regressor, raw features, testing the code with K = 4
All_X_70_nc = All_X_70.copy(deep=True)
All_X_70_nc = All_X_70_nc.drop(columns=['const'])
All_X_30_nc = All_X_30.copy(deep=True)
All_X_30_nc = All_X_30_nc.drop(columns=['const'])
#raw_neighbors = KNeighborsRegressor(n_neighbors = 4)
#raw_neighbors.fit(All_X_70_nc, PE_70)
#raw_neighbors.score(All_X_70_nc, PE_70)
```

```python
#Raw Train Error vs K
k_values = np.arange(1,101,1)
k_inv = 1/k_values
k_train_output = []
for Q in k_values:
    raw_scan = KNeighborsRegressor(n_neighbors=Q)
    raw_scan.fit(All_X_70_nc, PE_70)
    z = raw_scan.score(All_X_70_nc, PE_70)
    k_train_output.append(z)
```

```python
plt.plot(k_inv, k_train_output)
plt.title('Raw Data K Scan vs Train Data')
plt.xlabel("1/k")
plt.ylabel("R^2 against train data")
```

```python
#Raw Test Error vs K
k_test_output = []
for Q in k_values:
    raw_scan2 = KNeighborsRegressor(n_neighbors=Q)
    raw_scan2.fit(All_X_70_nc, PE_70)
    z = raw_scan2.score(All_X_30_nc, PE_30)
    k_test_output.append(z)
```

The value of K that gives me the best fit against the raw training data is 1.

```python
plt.plot(k_inv, k_test_output)
plt.title('Raw Data K Scan vs Test Data')
plt.xlabel("1/k")
plt.ylabel("R^2 against test data")
```

The value of K that gives me the best against the raw test data is 6.

```python
#KNN regressor, normalized features, testing the code with K = 4
#First I create test and train sets of the normalized data
All_X_norm_70 = All_X_norm.sample(n=dflen_s, random_state=randomState)
All_X_norm_30 = All_X_norm.drop(All_X_norm_70.index)
#now I make the model
#norm_neighbors = KNeighborsRegressor(n_neighbors = 4)
#norm_neighbors.fit(All_X_norm_70, PE_70)
```

Question: Indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.  
(a) The sample size n is extremely large, and the number of predictors p is small.  
(b) The number of predictors p is extremely large, and the number of observations n is small.  
(c) The relationship between the predictors and response is highly non-linear.  
(d) The variance of the error terms, i.e. σ2 (sigma squared) = Var(epsilon), is extremely high.  

__(a)__  A flexible method is preferred here because the size of the sample data will protect the model from over-fitting.  
__(b)__  A more inflexible method is preferred here because the small sample size will cause an overly flexible model to over-fit to the error in these data points, causing variance.  
__(c)__  A more flexible method is preferred in this case, because flexible models are created by estimating a greater number of parameters, and a non-linear relationship between a predictor and a response requires more parameters to correctly model it.  
__(d)__  A non-flexible method is preferred in this case, as a more flexible method would over-fit to these points which are skewed by irreducible error.
